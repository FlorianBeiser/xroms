# How to set up grid

The package `xcgm` has many nice, related functions for ROMS users. However, a bit of set up is required to connect from ROMS to the expected MITgcm output. This grid set up does that.

## Load in data

More information at in [load_data notebook](load_data.ipynb)

```python
loc = 'http://barataria.tamu.edu:8080/thredds/dodsC/forecast_latest/txla2_his_f_latest.nc'
chunks = {'ocean_time':1}
ds = xr.open_dataset(loc, chunks=chunks)
```

## Set up grid

```python
ds, grid = xroms.roms_dataset(ds)
```
